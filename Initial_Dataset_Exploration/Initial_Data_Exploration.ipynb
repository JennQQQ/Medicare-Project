```python
def save_each_year(path,file_name):
    col = df.columns.tolist()
    col = [c for c in col if c!='year']
    for year in range(2013,2024):
        temp = df[df['year']==year][col]
        temp.to_csv(f'{path}/{file_name}_{year}.csv')
    return
    
```

# Check all other columns that should be integers

```python
def check_non_integer_columns(percentage):
    cols = df.select_dtypes('number')

    for c in cols:
        mask_temp = (df[c]%1!=0)
        if df[mask_temp].shape[0]/df.shape[0]<percentage:
            df[mask_temp]=round(df[mask_temp])
    return df
        
```
