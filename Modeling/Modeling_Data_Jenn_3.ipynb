```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

pca_cols = ['Tot_Sbmtd_Chrg','Tot_Mdcr_Alowd_Amt']

# take the columns, coerce to numeric, handle missing
X = df[pca_cols].apply(pd.to_numeric, errors="coerce")
X = X.fillna(X.median(numeric_only=True))

# PCA is scale-sensitive -> standardize first
X_scaled = StandardScaler().fit_transform(X)

# 1 principal component -> one new column
pca = PCA(n_components=1, random_state=0)
df["pca_1"] = pca.fit_transform(X_scaled).ravel()

# (optional) see how much variance it explains and the loadings
print("Explained variance ratio:", pca.explained_variance_ratio_[0])
loadings = pd.Series(pca.components_[0], index=pca_cols).sort_values(key=abs, ascending=False)
print("PC1 loadings:\n", loadings)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(1, 10))
df["pca_1_scaled"] = scaler.fit_transform(df[["pca_1"]]).ravel()
```

# Label Data based on scatter plot of type and risk

```python
from sklearn.preprocessing import MinMaxScaler

def scattor_plots(col1,col2,dfR):
    d = dfR.dropna(subset=["Rndrng_Prvdr_Type", col2, f"{col1}"]).copy()
    scaler = MinMaxScaler()

    d[col2] = scaler.fit_transform(d[[col2]])
    d[f'{col1}_Scaled'] = scaler.fit_transform(d[[f'{col1}']])
    # (optional) if you have tons of types, keep top N by count
    TOP_N = dfR['Rndrng_Prvdr_Type'].nunique()
    types = d["Rndrng_Prvdr_Type"].value_counts().head(TOP_N).index.tolist()

    n = len(types)
    ncols = 4
    nrows = int(np.ceil(n / ncols))

    fig, axes = plt.subplots(nrows, ncols, figsize=(4*ncols, 4*nrows), sharex=True, sharey=True)
    axes = np.array(axes).ravel()

    for ax, t in zip(axes, types):
        sub = d[d["Rndrng_Prvdr_Type"] == t]
        ax.scatter(sub[col2], sub[f"{col1}_Scaled"], s=15, alpha=0.7)
        ax.set_title(str(t))
        ax.set_xlabel("Tot_Risk")
        ax.set_ylabel(f"{col1}")

    # turn off unused panels
    for ax in axes[len(types):]:
        ax.axis("off")

    plt.tight_layout()
    plt.show()
    return
```

# Birth clustering

```python
from sklearn.cluster import Birch
from sklearn.preprocessing import MinMaxScaler

def label_data(df,type,col1,col2):
    type_mask = df['Rndrng_Prvdr_Type']==type
    X = df[type_mask][[col1]+[col2]].copy()
    scaler = MinMaxScaler()
    X[f'scaled_{col1}'] = scaler.fit_transform(X[[col1]])
    X[f'scaled_{col2}'] = scaler.fit_transform(X[[col2]])
    X['Slope'] = (X[f'scaled_{col2}']/X[f'scaled_{col1}']).apply(lambda x:x if x<=120000 else 120000)
    #scaler = StandardScaler()
    #X_scaled = scaler.fit_transform(X[['slope']])

    if type=='LabPathology':
        birch = Birch(threshold=0.2, branching_factor=50, n_clusters=3)
        #labels = km.fit_predict(X_scaled)
        labels = birch.fit_predict(np.arctan(X[['Slope']]))
        df.loc[X.index,f'{type}_Tot_Risk'] = labels
        df[f'{type}_Tot_Risk'] = df[f'{type}_Tot_Risk'].fillna(3)
    else:
        birch = Birch(threshold=0.1, branching_factor=50, n_clusters=2)
        #labels = km.fit_predict(X_scaled)
        labels = birch.fit_predict(np.arctan(X[['Slope']]))
        df.loc[X.index,f'{type}_Tot_Risk'] = labels
        df[f'{type}_Tot_Risk'] = df[f'{type}_Tot_Risk'].fillna(2)

    
    return df
```

```python
temp = df.copy()

for t in select_type:
    label_data(temp,t,'Tot_Risk','Tot_Mdcr_Pymt_Amt')

temp
```

```python
fig, axes = plt.subplots(2,3,figsize=(15,10))

for i in range(0,5):
    t = select_type[i]
    mask_type = temp['Rndrng_Prvdr_Type']==t
    axes[i//3,i%3].scatter(temp[mask_type].loc[:,'Tot_Risk'],temp[mask_type].loc[:,'Tot_Mdcr_Pymt_Amt'],c=temp[mask_type][f'{t}_Tot_Risk'])
    axes[i//3,i%3].set_xlabel('Bene_Risk')
    axes[i//3,i%3].set_ylabel('Tot_Pymt')
    axes[i//3,i%3].set_title(f'{t}')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

NE = {"CT","ME","MA","NH","RI","VT","NJ","NY","PA"}
MW = {"IL","IN","MI","OH","WI","IA","KS","MN","MO","NE","ND","SD"}
S  = {"DE","DC","FL","GA","MD","NC","SC","VA","WV","AL","KY","MS","TN","AR","LA","OK","TX"}
W  = {"AZ","CO","ID","MT","NV","NM","UT","WY","AK","CA","HI","OR","WA"}

TERR = {"PR","VI","GU","AS","MP","PW","FM"}     # territories / compacts in your data
MIL  = {"AA","AE","AP"}                         # military “states”
UNK  = {"ZZ","XX"}                              # unknown / invalid catchalls (custom)

def group_geo(code: str) -> str:
    if pd.isna(code): return "Unknown"
    code = str(code).strip().upper()
    if code in NE:   return "Northeast"
    if code in MW:   return "Midwest"
    if code in S:    return "South"
    if code in W:    return "West"
    if code in TERR: return "Territory"
    if code in MIL:  return "Military"
    if code in UNK:  return "Unknown"
    return "Other/Unknown"

temp["state"] = temp["Rndrng_Prvdr_State_Abrvtn"].map(group_geo)
```

```python
def push_year(
        pre_df,
        curr_df,
        pre_cols = [c for c in cols if c not in ['Tot_Mdcr_Pymt_Amt','year','Is_Covid']],
        curr_cols = ['Tot_Mdcr_Pymt_Amt','year','Is_Covid']
    ): # everything but 'Tot_Mdcr_Pymt_Amt' are from previous year
    return pre_df[pre_cols].merge(curr_df[['Rndrng_NPI'] + curr_cols],on='Rndrng_NPI',how='inner')
```

```python
for i in range(2013,2023):
    mask_pre = df['year']==i
    pre_df = df[mask_pre]
    mask_curr = df['year']==i+1
    curr_df = df[mask_curr]
    if i==2013:
        common_NPI_pushed_year_df = push_year(pre_df,curr_df)
    else:
        common_NPI_pushed_year_df = pd.concat([common_NPI_pushed_year_df,push_year(pre_df,curr_df)],axis=0,ignore_index=True)

common_NPI_pushed_year_df
```
