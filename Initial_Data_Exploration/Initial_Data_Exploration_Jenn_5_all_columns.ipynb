## Missing BH datas

```python
# check missing data percentage for each column in BH

def check_missing_data_percentage(df,cols):
    result = {}
    total_mask = np.array([True] * df.shape[0]) # initial booleans for each row, assuming all rows are not nan
    for col in cols:
        mask = ~df[col].isna() # if a row is nan, then F; if a row is not nan, then T
        total_mask = (total_mask & mask) # if a row is nan, the row turns F; if a row is not nan, the row remains F
        result[col] = df[mask].shape[0]/df.shape[0]
    result['Good_Data_Percent'] = sum(total_mask)/df.shape[0]
    return pd.DataFrame([result],index=['nan_percentage']), total_mask

check_missing_data_percentage(dfBH1723,colsBH)[0]
```

## Number of beneficiaries by BH condition distribution

```python
# get the number of benes on only the good data rows
mask = check_missing_data_percentage(dfBH1723,colsBH)[1]
dfNumberBene = pd.DataFrame()

for col in colsBH:
    dfNumberBene[f'{col[:-7]}'] = round(dfBH1723[mask][col] * dfBH1723[mask]['Tot_Benes']/100)
    dfNumberBene[f'{col[:-7]}'] = dfNumberBene.loc[:,f'{col[:-7]}'].apply(lambda x: x if x!=0 else 1) # if it's zero, we make it to be 1 since we will take log

dfNumberBene
```

```python
# the histogram of log(number of patients)
np.log(dfNumberBene).hist(figsize=(16,12), bins=10)
```

## Assume missing data are less than 11 benes and impute with 5.5 benes

```python
# assume 5.5 benes for missing percentage data

def impute_missing_data(df,cols):
    for col in cols:
        mask = df[col].isna()
        temp = df[mask]
        temp.loc[:,col] = round(5.5/temp.loc[:,'Tot_Benes']*100)
        df.loc[mask,col]=temp[col]
    return df

dfimputeBH = dfBH1723.copy()
dfimputeBH = impute_missing_data(dfimputeBH,colsBH)
dfimputeBH
```

```python
dfNumberBene = pd.DataFrame()

for col in colsBH:
    dfNumberBene[f'{col[:-7]}'] = round(dfimputeBH[mask][col] * dfimputeBH[mask]['Tot_Benes']/100)
    dfNumberBene[f'{col[:-7]}'] = dfNumberBene.loc[:,f'{col[:-7]}'].apply(lambda x: x if x!=0 else 1) # if it's zero, we make it to be 1 since we will take log

np.log(dfNumberBene).hist(figsize=(16,12), bins=30)
```

## KMean cluster on the BH data

```python
from sklearn.cluster import KMeans

Ks = range(2, 11)
inertias = []

for k in Ks:
    km = KMeans(n_clusters=k, n_init=10, random_state=42)
    km.fit(dfimputeBH[colsBH])
    inertias.append(km.inertia_) # km.insertia_ = within-cluster sum of squares

plt.plot(list(Ks), inertias, marker="o")
plt.xlabel("k")
plt.ylabel("Inertia (within-cluster SSE)")
plt.title("Elbow plot for K-Means")
plt.show()
```

## Hierarchical Cluster on BH - can't be done due to too many data (more than a million)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering

X = dfBH1723[colsBH].fillna(0).to_numpy()
print("X shape:", X.shape)
X = StandardScaler().fit_transform(X)

# linkage: 'ward' (good default), or 'average', 'complete'

model = AgglomerativeClustering(n_clusters=4, linkage="ward")  # ward needs euclidean
labels = model.fit_predict(X)

dfBH1723["BH_label_HC_5"] = labels
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import MiniBatchKMeans
from scipy.cluster.hierarchy import linkage, fcluster
import matplotlib.pyplot as plt

num = dfBH1723[colsBH].fillna(0)
X = StandardScaler().fit_transform(num.to_numpy())

k = 500  # choose 100–2000
km = MiniBatchKMeans(n_clusters=k, batch_size=10000, random_state=0)
km.fit(X)

centers = km.cluster_centers_

# linkage: 'ward' (good default), or 'average', 'complete'
Z = linkage(centers, method="ward") # ward uses Euclidean distance

# Cut the tree into exactly 5 final clusters
final_k = 5
center_final_label = fcluster(Z, t=final_k, criterion="maxclust")  # length k, labels 1..5

# 6) Assign final cluster to each training row
micro_label = km.predict(X)                 # 0..k-1
final_label = center_final_label[micro_label]  # 1..5

dfBH1723['BH_HC_label'] = final_label
dfBH1723
```

### Boxplot for each cluster
