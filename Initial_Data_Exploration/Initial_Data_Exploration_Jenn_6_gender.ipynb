# Gender columns

```python
colsG = ['Rndrng_NPI','Rndrng_Prvdr_Type','Tot_Benes','Tot_Srvcs','Tot_Mdcr_Pymt_Amt','Bene_Feml_Cnt','Bene_Male_Cnt']
dfG = df[colsG]
dfG
```

## Nan rows

```python
pd.set_option("display.float_format", "{:,.6f}".format)   # choose decimals
```

```python
mask = dfG['Bene_Feml_Cnt'].isna()
dfna = dfG[mask]
dfna['Tot_Benes'].describe(include='all')
```

**The majority of missing gender data less than 25 Tot_benes**

## Rows less than Tot_benes = 100

```python
mask100 = dfna['Tot_Benes']<=100
df100 = dfna[~mask100]
df100
```

```python
# check their count in each type
number_by_type = df100.groupby('Rndrng_Prvdr_Type').size().sort_values(ascending=False)
number_by_type
```

**Check on top 5 types**

```python
top5 = ['RadiologyImaging','SurgeryOther','OBGYN','APP','OncologyHeme']
dfG5 = dfG[dfG['Rndrng_Prvdr_Type'].isin(top5)].copy()
dfG5
```

```python
temp = dfG5[~dfG5['Bene_Feml_Cnt'].isna()].copy()
temp['Feml_Pct'] = round(temp['Bene_Feml_Cnt']/temp['Tot_Benes']*100)
temp['Male_Pct'] = round(temp['Bene_Male_Cnt']/temp['Tot_Benes']*100)
temp['More_Feml'] = temp['Bene_Feml_Cnt']>=temp['Bene_Male_Cnt']
temp.groupby('Rndrng_Prvdr_Type').agg(
    Feml_Pct_mean = ('Feml_Pct','mean'),
    Male_Pct_mean = ('Male_Pct','mean'),
    More_Feml = ('More_Feml','mean')
)
```

## Inspect All Good Rows

```python
mask = dfG['Bene_Feml_Cnt'].isna()
dfGG = dfG[~mask].copy()
dfGG['Tot_Mdcr_Pymt_Amt'] = dfGG['Tot_Mdcr_Pymt_Amt'].apply(lambda x:x if x!=0 else 1) # preventing -inf when take log
dfGG['Feml_Pct'] = round(dfGG['Bene_Feml_Cnt']/dfGG['Tot_Benes']*100)
dfGG['Male_Pct'] = round(dfGG['Bene_Male_Cnt']/dfGG['Tot_Benes']*100)
dfGG['Log_Pymt_Amt'] = np.log(dfGG['Tot_Mdcr_Pymt_Amt'])
dfGG
```

```python
numcols = dfGG.select_dtypes(include='number').columns
corr = dfGG[numcols].corr()

plt.figure(figsize=(16,12))
sns.heatmap(corr,annot=True,cmap='RdBu_r',center=0)
plt.show()
```

**The PCT has some predictive power, it is resonable that the corr for Feml and Male are opposite**

```python
dfGG['FvsM'] = dfGG['Bene_Feml_Cnt']/dfGG['Bene_Male_Cnt']
dfGG['MvcF'] = dfGG['Bene_Male_Cnt']/dfGG['Bene_Feml_Cnt']
numcols = dfGG.select_dtypes(include='number').columns
corr = dfGG[numcols].corr()

plt.figure(figsize=(16,12))
sns.heatmap(corr,annot=True,cmap='RdBu_r',center=0)
plt.show()
```
