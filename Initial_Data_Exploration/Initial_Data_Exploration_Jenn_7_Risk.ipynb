# Risk Column

```python
colsR = ['Rndrng_NPI','Rndrng_Prvdr_Type','Tot_Benes','Tot_Srvcs','Tot_Mdcr_Pymt_Amt','Bene_Avg_Risk_Scre']
dfR = df[colsR].copy()
dfR
```

```python
dfR['Tot_Mdcr_Pymt_Amt'] = dfR['Tot_Mdcr_Pymt_Amt'].apply(lambda x:x if x!=0 else 1) # preventing -inf when take log
dfR['Log_Pymt_Amt'] = np.log(dfR['Tot_Mdcr_Pymt_Amt'])
dfR['Pymt_per_Srvcs'] = dfR['Tot_Mdcr_Pymt_Amt'] / dfR['Tot_Srvcs']
dfR['Pymt_per_Bene'] = dfR['Tot_Mdcr_Pymt_Amt'] / dfR['Tot_Benes']

numcols = dfR.select_dtypes(include='number').columns
numcols = [c for c in numcols if c!='Rndrng_NPI']
corr = dfR[numcols].corr()

plt.figure(figsize=(16,12))
sns.heatmap(corr,annot=True,cmap='RdBu_r',center=0)
plt.show()
```

```python
## Box plot of risk for each type

d = dfR.dropna(subset=["Rndrng_Prvdr_Type", "Bene_Avg_Risk_Scre"]).copy()

# optional: order types by median risk score
order = d.groupby("Rndrng_Prvdr_Type")["Bene_Avg_Risk_Scre"].median().sort_values().index

data = [d.loc[d["Rndrng_Prvdr_Type"] == t, "Bene_Avg_Risk_Scre"].values for t in order]

plt.figure(figsize=(12, 6))
plt.boxplot(data, labels=order, showmeans=True)
plt.xticks(rotation=45, ha="right")
plt.xlabel("Rndrng_Prvdr_Type")
plt.ylabel("Bene_Avg_Risk_Scre")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.patches as mpatches

# Clean
d = dfR.dropna(subset=["Rndrng_Prvdr_Type", "Bene_Avg_Risk_Scre", "Log_Pymt_Amt"]).copy()

# Optional: limit to top N provider types (otherwise the grid can get huge)
TOP_N = dfR['Rndrng_Prvdr_Type'].nunique()
types = d["Rndrng_Prvdr_Type"].value_counts().head(TOP_N).index.tolist()

n = len(types)
ncols = 4
nrows = int(np.ceil(n / ncols))

fig, axes = plt.subplots(nrows, ncols, figsize=(4*ncols, 3.5*nrows), sharey=True)
axes = np.array(axes).ravel()

# Colors (you asked for different colors)
c1, c2 = "tab:blue", "tab:orange"

for ax, t in zip(axes, types):
    sub = d[d["Rndrng_Prvdr_Type"] == t]

    x1, x2 = 1.00, 1.18  # close positions -> overlapping feel

    bp1 = ax.boxplot(
        sub["Bene_Avg_Risk_Scre"].values,
        positions=[x1],
        widths=0.28,
        patch_artist=True,
        showmeans=True
    )
    bp2 = ax.boxplot(
        sub["Log_Pymt_Amt"].values,
        positions=[x2],
        widths=0.28,
        patch_artist=True,
        showmeans=True
    )

    # Color + transparency so overlap is visible
    bp1["boxes"][0].set_facecolor(c1); bp1["boxes"][0].set_alpha(0.45)
    bp2["boxes"][0].set_facecolor(c2); bp2["boxes"][0].set_alpha(0.45)

    ax.set_title(str(t))
    ax.set_xticks([x1, x2])
    ax.set_xticklabels(["Risk", "LogPay"], rotation=0)

# Turn off unused panels
for ax in axes[len(types):]:
    ax.axis("off")

# Legend
handles = [
    mpatches.Patch(color=c1, alpha=0.45, label="Bene_Avg_Risk_Scre"),
    mpatches.Patch(color=c2, alpha=0.45, label="Log_Pymt_Amt"),
]
fig.legend(handles=handles, loc="upper center", ncol=2)

plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()
```

```python
def scattor_plots(col):
    d = dfR.dropna(subset=["Rndrng_Prvdr_Type", "Bene_Avg_Risk_Scre", f"{col}"]).copy()

    # (optional) if you have tons of types, keep top N by count
    TOP_N = dfR['Rndrng_Prvdr_Type'].nunique()
    types = d["Rndrng_Prvdr_Type"].value_counts().head(TOP_N).index.tolist()

    n = len(types)
    ncols = 4
    nrows = int(np.ceil(n / ncols))

    fig, axes = plt.subplots(nrows, ncols, figsize=(4*ncols, 4*nrows), sharex=True, sharey=True)
    axes = np.array(axes).ravel()

    for ax, t in zip(axes, types):
        sub = d[d["Rndrng_Prvdr_Type"] == t]
        ax.scatter(sub["Bene_Avg_Risk_Scre"], sub[f"{col}"], s=15, alpha=0.7)
        ax.set_title(str(t))
        ax.set_xlabel("Bene_Avg_Risk_Scre")
        ax.set_ylabel(f"{col}")

    # turn off unused panels
    for ax in axes[len(types):]:
        ax.axis("off")

    plt.tight_layout()
    plt.show()
    return
```

```python
scattor_plots('Log_Pymt_Amt')
```
